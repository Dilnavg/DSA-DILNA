```python
plt.figure(figsize= (10,6))
sns.countplot(data['parental level of education'])
plt.title('students parental level of education')
plt.xticks(rotation=90)
plt.ylabel('Number of students')
```

```python
plt.figure(figsize= (10,6))
data['Total Score']=data['math score']+data['reading score']+data['writing score']
plt.title("Total Score v/s parental level of education")
sns.barplot(x=data['parental level of education'],y=data['Total Score'])
```

```python
    ### Students having parents  with  masters degree or bachelors degree improves the total score in students perfomance.
    #Parents with The high school background showing less total score than others with masters or bachelors  
```

```python
##ANSWER 3:AVERAGE ON MATHS
```

```python
fig,ax = plt.subplots(3,1, figsize = (5,10))
sns.boxplot(x=data['gender'],y=data['math score'],ax=ax[0])
sns.boxplot(x=data['gender'],y=data['reading score'],ax=ax[1])
sns.boxplot(x=data['gender'],y=data['writing score'],ax=ax[2])
plt.tight_layout()
```
